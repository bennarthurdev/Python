```python
#Series é um objeto do tipo array unidimensional contendo uma sequência de valores e um array de rótulos denominado índice. 

from pandas import Series

array = ['Julie' , 'Stwart' , 'Mozart']
#o método Series por si só, não modifica o objeto array, então é nescessário salvar o objeto do tipo Series em uma nova variável.
array_series = Series(array)
array_series
```

```python
#Series é também um dicionário ordenado de tamanho fixo, como se fosse um mapemaneto entre índices e valores de dados.
'Julie' in array_series
```
