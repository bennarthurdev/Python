## Universal Functions
 - functions to realize operations for all elements in data of ndarrays
  - ufuncs can be unary, if your identifier list receive just one array
  - ufuncs are binary if receive two arrays as a param

````python
arr6 = np.random.randn(10)
arr6      
````

````python
arr7 = np.random.randn(10)
arr7
````

````python
#Returns the max between arr6 and arr7 for all elements
np.maximum(arr6 , arr7)
````

````python
#float and integer parts of array elements in a tuple with float point notation
arr8 = np.random.randn(10)
print(arr8)
arr8 = arr8 * 2 # Example of unary ufunc
print(arr8) 
float_part , int_part = np.modf(arr8)
print(float_part)
print(int_part)
````

## Conditional Logic 

```
# numpy.where(
   c -> boolean condition or array
   x -> array who receives values if c is True
   y -> array who receives values if c is False
  )
```




````python
#generate alleatory array
arr9 = np.random.randn(4,4)
#generate mask
arr10 = np.zeros((4,4))
arr10[:,:] = 0.5
#create a condition
c = arr9 < 0.5
#call method
np.where(c , arr9 , arr10)

````

## Exploratory Data Analysis
### Dispersion measures
#### Variance
####varianciadaamostra.webp

````python
set_1 = [-10,0,10,20,30]
set_1_ndarray = np.array(set_1)
def varianceRoot(serie , sum = 0):
  #mdia aritmtica da srie
  x = np.mean(serie) 
  #somatrio
  for xi in serie:
    sum += (xi-x)**2
  #varincia
  return (sum/len(serie)) 
varianceRoot(set_1)
````

````python
variance = np.var(set_1)
````

### Standard Deviation
The standard Deviation is the square root of variance, with stantard unity of measure

It is a widely used parameter in statistics that indicates the degree of variation of a set of elementsdesvio-padrao-matematica.webp

````python
Dp = math.sqrt(variance)
Dp
````
