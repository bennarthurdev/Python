## Métodos Preditivos
- Classificação 
  - Define classes para cada um dos registros (rótulos)
  - Exemplos:
    - Identificar potencial de compra para terminado produto com base no comportamento do cliente, pesquisas de insatisfação do cliente, **Risco de crédito**, filtros de SPAM, separação de notícias, reconhecimento de voz, reconhecimento de face, previsão de doenças.

- Regressão
  - Prever valor de vendas, comportamentos atmosféricos, probabilidade de um paciente sobreviver, riscos de investimento, gastos no cartão de crédito(limites automáticos).

## Métodos Descritivos 

### Associação 
- Prateleiras de Mercado (fraldas e cervejas)

- Planejar catálogos das lojas e folhetos de promoções
- Controle de evasão em universidades

### Agrupamento - agrupar por meio da observação de padrões de um objeto

- Encontrar grupos de clientes que irão comprar um produto(mala direta)
- Agrupamento de documentos/notícias
- Agrupamento de produtos similares
- Perfis de clientes(Netflix)
- Análise de redes sociais

### Detecção de desvios (outliers) - foge do padrão
- Fraude em cartão de crédito
- Intrusão em redes 
- Uso de energia elétrica, água ou telefone
- Doping
- Monitorar máquinas em um data center

### Descoberta de padrões sequenciais

## Aprendizagem de Máquina Supervisionada

### Aprendizagem supervisionada - Fase 1
1. Extração de características (montagem da base de dados)
2. Algoritmo de aprendizagem (supervisor)
3. Modelo aprendido (conhecimento extraído pelo algoritmo)

### Aprendizagem supervisionada - Fase 2
1. Extração de características 
2. Modelo aprendido




## Aprendizagem de Máquina Não Supervisionada
- Analisar automaticamente os dados (associação, agrupamento)
- Necessita análise para determinar o significado dos padrões encontrados.

## Aprendizagem por reforço
- Aprender com as interações com o ambiente (causa e efeito)
- Aprender com a sua própria experiência 
- Exemplo: Robô que aprende a coletar lixo em um ambiente -> aplicação em sistemas multi-agentes.

- Objetiva-se descobrir um relacionamento entre os atributos previsores e o atributo meta
- O valor do atributo meta é conhecido (aprendizagem supervisionada)
