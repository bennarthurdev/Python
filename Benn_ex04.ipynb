### Qual o gênero de filme com maior nota do IMDB? E o menor?

```python
#modules
import numpy as np
import pandas as pd

#import dataset from github
URL = "https://github.com/yurimalheiros/icd/blob/master/datasets/movies.csv?raw=true" 

#create dataframe
df = pd.read_csv(URL)
#see dataframe
df
```

# Funções


*   defineGenres:<br>
    Returns the genres contained in the movie.<br>
    Receives an array and iterates over it <br>
    if the element is already present, that type is already defined <br>
    otherwise that element can be NaN, as pandas will see it as float64<br>
    if the element is not present in the array "types" then it is defined and included in the array
  
    
    
*   Preprocessing : <br>
    args : dataFrame and Column Name<br>
    1.Eliminate NAN or NUll in specif column of an dataframe<br>
    2.Don't modify the original object, but return a new object who can be copied 
 


```python

def defineGenres(array , types = list(),i = 0):

  for x in array:
    
    if x in types:
      pass

    elif ( type(array[i]) == float ):
      pass

    else:
      types.append(array[i])
      
    i += 1

  return(types)
```

```python
def Preprocessing(dataFrame , columns):

  for column in columns:
    df = dataFrame.dropna(subset = column)
  return df
```

1 - Excluir Valores indesejados
2 - Vizualizar os gêneros de filme presentes

```python

df = Preprocessing(df, [['IMDB Rating'] , ['Major Genre']])
 
tipos = defineGenres(df["Major Genre"].values)

tipos
```

```python
genres = df['Major Genre'] #Series with Major Genre can be acessed in this form

df[genres.values == tipos[0]]['IMDB Rating'].mean() #Calculate the mean for the first element in "tipos"
```

```python
#But I need all the averages for all the film types
means = list()
for i in range(0, 12 , 1):
  means.append(df[genres.values == tipos[i]]['IMDB Rating'].mean())
means
```

```python
means_df = pd.DataFrame(means , columns = ['means'] , index = tipos) #Create a dataFrame with means of Filmes
means_df
```

```python
#OUTPUT
means_df.sort_values(by = 'means'  , ascending=False)
```

### Qual o diretor dirigiu mais filmes de drama? E de comédia?

```python
df
```

```python
#Filtrar por Major Genre

df_drama = df[df['Major Genre'] == 'Drama'] #seleciona as linhas para as quais a operação df['Major Genre'] == 'Drama' recebe true
grouped_drama = df_drama['Director'].value_counts()

df_comedy = df[df['Major Genre'] == 'Comedy']
grouped_comedy = df_comedy['Director'].value_counts()

print(f"DRAMA \n{grouped_drama}\n\n\n\nCOMÉDIA \n{grouped_comedy}")
```
